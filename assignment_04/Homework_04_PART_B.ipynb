```python
# use apply function to cleanse data

def cleanse_data(values):
    data = values.replace("*","")
    return data
```

```python
df["year"] = df["Census Year"].apply(cleanse_data)
df['year'] = pd.to_numeric(df['year'])
df
```

```python
def cleanse_data(values):
    data = values.replace(",","")
    return data
```

```python
df["population"] = df["Total Population"].apply(cleanse_data)
df['population'] = pd.to_numeric(df['population'])
df
```

```python
def cleanse_data(value):
  data = value.replace(",","").replace("$","").replace(" Billion","000000000").replace("cents","").replace("\t","")
  return data
```

```python
df["cost_census"] = df["Census Cost"].apply(cleanse_data)
df["avg_person"] = df["Average Cost Per Person"].apply(cleanse_data)
df
```

```python
df['cost_census'] = pd.to_numeric(df['cost_census'])
df['avg_person'] = pd.to_numeric(df['avg_person'])
```
